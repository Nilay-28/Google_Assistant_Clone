# **DEPENDENCIES**

```python
#Import Necessary Dependencies
!pip install -q -U google-generativeai
!pip install gTTS
import pathlib
import textwrap
import google.generativeai as genai
from google.colab import userdata
from gtts import gTTS
from IPython.display import display, Audio
from IPython.display import Markdown

#Default Function By Google Gemini To Format Ouptut Generated By AI ChatBot.
def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

# **LOAD GEMINI**

```python
#Load Gemini-Pro Model And Load The API_KEY To The Model
Google_Gemini_API=userdata.get('Google_Gemini_API')
genai.configure(api_key=Google_Gemini_API)
model = genai.GenerativeModel('gemini-pro')
```

# **FUNCTIONS**

```python
def Intro(name):
  print(f"Hey {name}!")

def Intro2():
  print("How Can I Help You? (Type stop To Exit)")

def Questions(ques):
  device_list = ['alarm', 'timer', 'message', 'call']
  flag = False

  for i in device_list:
    if i in ques.lower():
      flag = True
      break

  if flag == True:
    print("Not Supported Yet.\nWe Will Be Updating This Assistant Soon!!\nSTAY TUNED :)")
    return("Not Supported Yet.\nWe Will Be Updating This Assistant Soon!!\nSTAY TUNED :)")
    flag = False

  if flag == False:
    if "who created you" in ques.lower():
      print("Nilay Koul Created Me")
      return("Nilay Koul Created Me")

    elif "what is your name" in ques.lower() or "who are you" in ques.lower() or "your name?" in ques.lower():
      print("I am a Google Assistant Clone\nYou Can Ask Me Any Questions!")
      return("I am a Google Assistant Clone\nYou Can Ask Me Any Questions!")

    else:
      prompt = ques + " in 20 words"
      response = model.generate_content(prompt)
      text1 = to_markdown(response.text)
      display(text1)
      return text1.data

def Text_To_Speech(ques):
  audio1 = gTTS(ques)
  audio1.save("audio1.mp3")
  return display(Audio("audio1.mp3", autoplay=True))
```

# **MAIN**

```python
def Assistant():
  name = input("Hey!, What Is Your Name?\n")
  Intro(name)
  STOP = True
  while STOP:
    Intro2()
    ques=(input())
    if "stop" in ques.lower():
      STOP = False
      break
    else:
      Text_To_Speech(Questions(ques))

Assistant()
```
